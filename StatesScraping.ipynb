# Web Scraping - Wikipedia Link for population in states of USA

```python
#Build in libraries for request, Beautiful Soup and Dataframe
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import requests  
import pandas as pd


#Providing the URL from which web scraping will be done
url = 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population'
response =  requests.get(url)

#For parsing using Beautiful soup
page_soup = soup(response.text, "html.parser")

#Scraping of the table 
table = page_soup.find("table",{"class":"wikitable sortable"}).tbody
#table = containers.text
#print (table)

#scraping of all the rows of the table in wikipedia link
rows = table.findAll("tr")
#print(rows)
    
#scraping headers of table     
columns = [v.text.replace('\n',"") for v in rows[0].find_all("th")]
#print (columns)

#storing headers in dataframe 
df = pd.DataFrame(columns=columns)

#loop for the all the rows data
for i in range(1, len(rows)):
    tds = rows[i].find_all("td")
    #print (tds)
   
    #replacing junk values with blank and storing data in pandas dataframe   
    values = [tds[0].text.replace("\n",""),tds[1].text.replace("\n","").replace("[d]",""), tds[2].text.replace("\n","").replace("\xa0",""), tds[3].text.replace("\n",""), tds[4].text.replace("\n","").replace("Ã¢Ë†â€™",""), tds[5].text.replace("\n","").replace("Ã¢Ë†â€™",""), tds[6].text.replace("\xa0sq\xa0mi\n","").replace("Ã¢Ë†â€™",""), tds[7].text.replace("\xa0km2\n",""),tds[8].text.replace("\xa0mi\n","")]
        
    #print (values)
  
    #appending values with columns in pandas dataframe and printing the same 
    df = df.append(pd.Series(values, index=columns),ignore_index = True)
    print (df)
```

```python
#converting data frame to a CSV file
df.to_csv("statesWiki.csv", encoding='utf-8', index=False)
```
